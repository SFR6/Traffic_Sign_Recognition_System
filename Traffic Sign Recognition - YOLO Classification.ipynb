```python
from ultralytics import YOLO

model = YOLO("yolo11n-cls.pt")
```

# Hyperparameter Tuning

```python
import itertools

params = {
    "imgsz": [512, 256, 128],
    "batch": [8, 16, 32],
    "lrf": [0.005, 0.001],
    "cos_lr": [False, True],
    "augment": [False, True],
}

augmentation_params = {
    "hsv_h": 0.015, "hsv_s": 0.7, "hsv_v": 0.4,
    "degrees": 10, "translate": 0.1, "scale": 0.5, "perspective": 0.1,
}

param_combinations = list(itertools.product(*params.values()))

for combination in param_combinations:
    param_dict = dict(zip(params.keys(), combination))
    
    print(f"Training with parameters: {param_dict}")

    if param_dict["augment"]:
        param_dict.update(augmentation_params)

    model.train(
        data="C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/traffic_signs",
        imgsz=param_dict["imgsz"],
        batch=param_dict["batch"],
        lrf=param_dict["lrf"],
        cos_lr=param_dict["cos_lr"],
        augment=param_dict["augment"],
        epochs=200,
        device=1,
        patience=50,
        **{key: param_dict[key] for key in augmentation_params.keys() if key in param_dict}
    )
```

# Confidence Threshold Tuning

#### WITHOUT Assuming Flawless Detection

```python
import os
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from ultralytics import YOLO

meta_data_test_2 = pd.read_csv("C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/MyTest2.csv")
initial_path = "C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/"
image_paths = meta_data_test_2["Path"]
label_array = meta_data_test_2["ClassId"]

confidence_thresholds = [0.6, 0.7, 0.8, 0.9, 0.91, 0.92, 0.93, 0.94, 0.95, 0.96, 0.97, 0.98, 0.99]

base_path = "C:/Users/Tudor Cristea/Documents/IP/project/venv_PRS/Scripts/runs/classify/"
model_paths = []

for i in range(32):
    train_folder = "train" + "2" * i
    model_path = os.path.join(base_path, train_folder, "weights", "best.pt")
    if os.path.exists(model_path):
        model_paths.append(model_path)

for i in range(32):
    train_folder = "train3" + "2" * i
    model_path = os.path.join(base_path, train_folder, "weights", "best.pt")
    if os.path.exists(model_path):
        model_paths.append(model_path)

csv_path = "model_evaluation_results.csv"
results = []

if os.path.exists(csv_path):
    existing_results_df = pd.read_csv(csv_path)
    existing_models = existing_results_df["model"].unique()
else:
    existing_results_df = pd.DataFrame()
    existing_models = []

for model_path in model_paths:

    if model_path in existing_models:
        print(f"Skipping already evaluated model: {model_path}")
        continue

    print(f"Evaluating model: {model_path}")
    model = YOLO(model_path)
    
    for threshold in confidence_thresholds:
        label_pred_array = []
        final_label_array = []

        for image_path, label in zip(image_paths, label_array):
            file_path = initial_path + image_path
            result = model(file_path, verbose=False)
            pred = result[0].names[result[0].probs.top5[0]]
            confidence = result[0].probs.top5conf[0].item()
            if label != -1:
                final_label_array.append(label)
                if confidence > threshold:
                    label_pred_array.append(int(pred))
                else:
                    label_pred_array.append(-1)
            else:
                if confidence > threshold:
                    final_label_array.append(label)
                    label_pred_array.append(int(pred))

        if final_label_array and label_pred_array:
            accuracy = accuracy_score(final_label_array, label_pred_array)
            precision = precision_score(final_label_array, label_pred_array, average='macro', zero_division=0)
            recall = recall_score(final_label_array, label_pred_array, average='macro', zero_division=0)
            f1 = f1_score(final_label_array, label_pred_array, average='macro')

            results.append({
                "model": model_path,
                "threshold": threshold,
                "accuracy": accuracy,
                "precision": precision,
                "recall": recall,
                "f1_score": f1
            })
        else:
            results.append({
                "model": model_path,
                "threshold": threshold,
                "accuracy": 0,
                "precision": 0,
                "recall": 0,
                "f1_score": 0
            })

new_results_df = pd.DataFrame(results)
if not existing_results_df.empty:
    combined_results_df = pd.concat([existing_results_df, new_results_df], ignore_index=True)
else:
    combined_results_df = new_results_df
combined_results_df.to_csv(csv_path, index=False)
```

# Best Models Selection

```python
import pandas as pd
import os
import yaml

def load_hyperparameters(model_path):
    train_folder = os.path.join(os.path.dirname(model_path), "..")
    args_file = os.path.join(train_folder, "args.yaml")
    if os.path.exists(args_file):
        with open(args_file, "r") as file:
            args = yaml.safe_load(file)
            if args.get("augment"):
                return {
                    "imgsz": args.get("imgsz"),
                    "batch": args.get("batch"),
                    "lrf": args.get("lrf"),
                    "cos_lr": args.get("cos_lr"),
                    "augment": args.get("augment"),
                    "hsv_h": args.get("hsv_h"), "hsv_s": args.get("hsv_s"), "hsv_v": args.get("hsv_v"),
                    "degrees": args.get("degrees"), "translate": args.get("translate"), "scale": args.get("scale"), "perspective": args.get("perspective"),
                }
            return {
                "imgsz": args.get("imgsz"),
                "batch": args.get("batch"),
                "lrf": args.get("lrf"),
                "cos_lr": args.get("cos_lr"),
                "augment": args.get("augment"),
            }
    return {}

results_df = pd.read_csv("model_evaluation_results.csv")
results_df["accuracy_f1_sum"] = results_df["accuracy"] + results_df["f1_score"]
results_df["accuracy_f1_diff"] = abs(results_df["accuracy"] - results_df["f1_score"])

best_model = results_df.sort_values(
    by=["accuracy_f1_sum", "accuracy_f1_diff"],
    ascending=[False, True]
).iloc[0]

hyperparameters = load_hyperparameters(best_model["model"])

print(f"Best Model:\n"
      f"Hyperparameters: {hyperparameters}\n"
      f"Threshold: {best_model['threshold']}\n"
      f"Accuracy: {best_model['accuracy']}\n"
      f"Precision: {best_model['precision']}\n"
      f"Recall: {best_model['recall']}\n"
      f"F1 Score: {best_model['f1_score']}\n")

results_df_perfect = pd.read_csv("model_evaluation_results_fd.csv")
results_df_perfect["accuracy_f1_sum"] = results_df_perfect["accuracy"] + results_df_perfect["f1_score"]
results_df_perfect["accuracy_f1_diff"] = abs(results_df_perfect["accuracy"] - results_df_perfect["f1_score"])

best_model_perfect = results_df_perfect.sort_values(
    by=["accuracy_f1_sum", "accuracy_f1_diff"],
    ascending=[False, True]
).iloc[0]

hyperparameters_perfect = load_hyperparameters(best_model_perfect["model"])

print(f"\nBest Model (assuming flawless detection):\n"
      f"Hyperparameters: {hyperparameters_perfect}\n"
      f"Threshold: {best_model_perfect['threshold']}\n"
      f"Accuracy: {best_model_perfect['accuracy']}\n"
      f"Precision: {best_model_perfect['precision']}\n"
      f"Recall: {best_model_perfect['recall']}\n"
      f"F1 Score: {best_model_perfect['f1_score']}\n")
```

# Final Image Creation

```python
import cv2
import pandas as pd
import os

def create_final_image(file_name_without_extension, label_pred_array, size=40):
    initial_image_path = "C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/MyTest/" + file_name_without_extension + ".png"
    intitial_image = cv2.imread(initial_image_path, cv2.IMREAD_COLOR)
    overlay_images_path = "C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/Meta/"
    result_image_path = "C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/MyTest/final_results_3/"
    height, width = intitial_image.shape[:2]
    if size * len(label_pred_array) < height:
        for i, label in enumerate(label_pred_array):
            overlay_image = cv2.imread(overlay_images_path + str(label) + ".png", cv2.IMREAD_COLOR)
            resized_overlay_image = cv2.resize(overlay_image, (size, size), interpolation = cv2.INTER_AREA)
            intitial_image[i * size:(i + 1) * size, width - size:width] = resized_overlay_image
        cv2.imwrite(result_image_path + file_name_without_extension + ".png", intitial_image)

meta_data_test_2 = pd.read_csv("C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/MyTest.csv")
initial_path = "C:/Users/Tudor Cristea/Documents/IP/project/datasets/archive/MyTest/result_images/"
images_paths = meta_data_test_2["Path"]

model = YOLO(best_model["model"])
confidence_threshold = best_model["threshold"]

for image_path in images_paths:
    file_name = os.path.splitext(os.path.basename(image_path))[0]
    file_name_without_extension = file_name.split('/')[-1]
    folder_path = initial_path + file_name_without_extension + "/"
    files = os.listdir(folder_path)

    label_pred_array = []
    for file in files:
        if os.path.isfile(folder_path + file):
            result = model(folder_path + file, verbose=False)
            pred = result[0].names[result[0].probs.top5[0]]
            confidence = result[0].probs.top5conf[0].item()
            if confidence > confidence_threshold and int(pred) not in label_pred_array:
                label_pred_array.append(int(pred))
        
    create_final_image(file_name_without_extension, label_pred_array)
```
